```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datapath = "../Dataset/Algerian_forest_fires_dataset_UPDATE.csv"
df = pd.read_csv(datapath,header=1)
df.head()
```

```python
df.info()
```

# Data Cleaning

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,'Region'] = 0
df.loc[122:,'Region'] = 1
df
```

```python
df.info()
```

```python
df.Region = df.Region.astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
columns = [col for col in df.columns if df[col].dtype =="object"]
columns
```

```python
for col in columns[:-1]:
    df[col] = df[col].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv("Algerian_forest_fires_dataset_cleaned.csv",index=False)
```

```python
dftemp = df_copy.loc[df.Region == 1]
plt.figure(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='Month',hue='Classes',data=df_copy)
plt.xlabel('Months',weight='bold')
plt.ylabel("Number of Fires",weight='bold')
plt.title("Monthly Fire Analysis of Sidi-Bel Regions",weight="bold")
```

```python
dftemp = df_copy.loc[df.Region == 0]
plt.figure(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='Month',hue='Classes',data=df_copy)
plt.xlabel('Months',weight='bold')
plt.ylabel("Number of Fires",weight='bold')
plt.title("Monthly Fire Analysis of Brjaia Regions",weight="bold")
```
